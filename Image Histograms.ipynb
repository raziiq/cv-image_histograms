# Image Histograms

### What are Histograms?

A histogram is a graphical representation that organizes a group of data points into user-specified ranges

#### Importing Required Libraries

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

#### Working with 1D Historgrams

```python
ages = np.random.randint(1, 100, size=1000)
plt.figure(figsize=(10,5))
plt.plot(ages, '.')
```

Numpy has a min and max function to provide min and max in a collection

```python
minimum = np.min(ages)
maximum = np.max(ages)

print("Min:", minimum, "Max:", maximum)
```

In order to draw a histogram, it requires you to provide it bins. We can calculate the bins using minimum and maximum values and dividing the bins into 50 equal slots

```python
nbins = 50

bins = np.linspace(minimum, maximum, nbins)
print(bins)
```

Numpy's histogram method helps in creating a histogram and returns frequency and bins information normally called edges

```python
hist, edges = np.histogram(ages, bins)
print("Hist:", hist, "Count:", len(hist))
print("Edges:", edges, "Count:", len(edges))
```

We can plot this using plt.plot method which would draw a scatterplot based on the frequencies.

```python
plt.plot(hist, 'r.', markersize=10)
```

Histograms are more meaningful when drawn using bar plots. For bar plots we need to provide the x and y axis. Y axis would be the count or frequency and X axis would be the bins. To make it clearer, it is a good idea to find the center of the bins to plot on X axis.

```python
centers = (edges[:-1] + edges[1:])/2
print(centers)
```

Plotting a histogram using bar plots in matplotlib.

```python
plt.figure(figsize=(10,8))
plt.bar(centers, hist, width=1.5)
```

Plotting a histogram using matplotlib's hist function. It requires you to provide edges having same count as the array of frequency of numbers, an array of bins and finally an arrays of frequency of numbers.

```python
plt.figure(figsize=(10,8))
plt.hist(edges[:-1], edges, weights=hist, rwidth=0.8)
```

```python
plt.figure(figsize=(10,8))
plt.hist(ages, bins, rwidth=0.8)
```

```python
plt.figure(figsize=(20,8))
plt.subplot(131)
plt.bar(centers, hist, width=1.5)

plt.subplot(132)
plt.hist(edges[:-1], edges, weights=hist, rwidth=0.8)

plt.subplot(133)
plt.hist(ages, bins, rwidth=0.8);
```

#### Working with 2D Historgrams

A 2D histogram shows the number of elements in each bin, by drawing bars of corresponding length

```python
x = [1.6, 2.5, 3.2, 1.2]
y = [1.3, 1.5, 2.3, 1.8]

bins = [1, 2, 3, 4]

plt.plot(x, y, '.', markersize = 10)
plt.xticks(bins)
plt.yticks(bins)
plt.grid()
```

```python
plt.hist2d(x, y, bins=bins)
plt.colorbar()
```

Creating normally distributed random numbers.

```python
x = np.random.randn(100000)
y = np.random.randn(100000)

plt.plot(x, y, '.')
```

```python
plt.figure(figsize=(12,8))
plt.hist2d(x, y, bins=100)
plt.colorbar()
```

#### Working with Image Historgrams

```python
image = cv.imread("lena.png")
image_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

hist, edges = np.histogram(image_gray, bins=256)

plt.figure(figsize=(12,8))

plt.subplot(121)
plt.imshow(image_gray, cmap="gray")

plt.subplot(122)
plt.hist(edges[:-1], edges, weights=hist);
```

There is a better way of dealing with image histograms using OpenCV

```python
plt.figure(figsize=(12,8))
hist = cv.calcHist([image_gray], [0], None, [256], [0,255])
plt.plot(hist)
```

#### Working with Color Image Historgrams

```python
image = cv.imread("lena.png")
b,g,r = cv.split(image)

plt.figure(figsize=(12, 8))
plt.subplot(121)
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))

plt.subplot(122)
hist = cv.calcHist([b], [0], None, [256], [0,255])
plt.plot(hist, color="blue")

hist = cv.calcHist([g], [0], None, [256], [0,255])
plt.plot(hist, color="green")

hist = cv.calcHist([r], [0], None, [256], [0,255])
plt.plot(hist, color="red")
```
